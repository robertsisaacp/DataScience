```python
# Import packages
import pandas as pd
import string
import re
import nltk
from sklearn.feature_extraction.text import CountVectorizer

stopword = nltk.corpus.stopwords.words('english')
wn = nltk.WordNetLemmatizer()
```

```python
# Read in data
pd.set_option('display.max_colwidth', 100)

data = pd.read_csv("SMSSpamCollection.tsv", sep='\t', header=None)
data.columns = ['label', 'body_text']

data.head()
```

```python
# Create utility function to clean text data in preparation for CountVectorizer with bigrams
def clean_text_bi(text):
    # strip the punctuation
    text_nopunct = "".join([char for char in text if char not in string.punctuation])
    # tokenize
    tokens = re.split('\W+', text_nopunct)
    # remove stop words and lemmatize
    text = " ".join([wn.lemmatize(word) for word in tokens if word not in stopword])
    return text

# Create utility function to clean text data in preparation for TF-IDF
def clean_text_idf(text):
    # strip the punctuation
    text_nopunct = "".join([char for char in text if char not in string.punctuation])
    # tokenize
    tokens = re.split('\W+', text_nopunct)
    # remove stop words and lemmatize
    text = [wn.lemmatize(word) for word in tokens if word not in stopword]
    return text
```

```python
data['clean_text_bi'] = data['body_text'].apply(lambda x: clean_text_bi(x.lower()))

# Apply CountVectorizer with bigrams
ngram_vect = CountVectorizer(ngram_range=(2,2))
X_counts = ngram_vect.fit_transform(data['clean_text_bi'])

# Store vectorized text in DataFrame
X_counts_df = pd.DataFrame(X_counts.toarray())
X_counts_df.columns = ngram_vect.get_feature_names()
X_counts_df
```

```python
# Apply TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer(analyzer=clean_text_idf)
x_tfidf = tfidf_vect.fit_transform(data['body_text'])

X_tfidf_df = pd.DataFrame(x_tfidf.toarray())
X_tfidf_df.columns = tfidf_vect.get_feature_names()
X_tfidf_df
```
