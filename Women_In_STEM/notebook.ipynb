# Women in STEM: A Look at Postsecondary Degrees Awarded

“Women make up half of the total U.S. college-educated workforce, but only 29% of the science and engineering workforce… (compared to men in the field) there are relatively low shares in engineering (15%) and computer and mathematical sciences (25%).”- National Girls Collaborative Project

Why care about female representation in STEM? 
[Bianca Barratt](https://www.forbes.com/sites/biancabarratt/2018/11/17/the-need-for-more-women-in-stem-roles-goes-beyond-simple-diversity/#38bd614445cb) states it beautifully and succintly:
"This push goes beyond the need for simple gender parity. It is not just a case of making the STEM workforce fair – we need more women in STEM roles to make scientific innovations useful and, more importantly, safe . After all, how relevant can innovations really be if they do not even take into consideration the needs of half the population?"

In this notebook, we will focus on exploring the statistics and trends on postsecondary degrees awarded to women between the years 1970 and 2011. 

![](img/women_in_stem.jpg)


For the purposes of this EDA, the following majors will be considered STEM: 
* Agriculture
* Architecture
* Biology
* Computer Science
* Engineering
* Health Professions
* Math and Statistics
* Physical Sciences
* Psychology

A full list of STEM degrees recognized by the Department of Homeland Security is available [here](https://www.ice.gov/sites/default/files/documents/Document/2014/stem-list.pdf).

### Objective and Questions to Answer
* Overall, has there been an increase or decrease in women getting postsecondary degrees?
* What were the most popular degrees for women in 1970, 1990 and 2011?
* Comparing STEM versus Non-STEM degrees: Is there a significant divide?
* Among the 17 types of degrees considered: Which ones have become more popular among women? Which ones are less popular than average?

### Key Takeaways:
* Health professions remains the most popular degree among women all through the years. Psychology entering the top 5 around 1990. 
* Health Professions, Psychology and Biology are consistently the top 3 STEM degrees for women.
* While computer science was the 6th most popular degree (13.6%), it is replaced by other STEM degrees (such as Math, Architecture and Agriculture) in subsequenct years. 
* Women now obtain postsecondary degrees at a higher rate compared to men. The percentage of total degrees awarded to women has grown from 38.6 to 56%.
* Women have had higher representation compared to men in Non-STEM fields since 1970 (52.9%). This statistic has increased to 65% over the years.
* It is also encouraging to see the close to 50% of STEM degrees are now awarded to women.
* However, when isolating Computer Science and Engineering degrees, women are still very underrepresented (about 18% respectively).
* We see a period of increased interest in computer science among women between the years 1970 and 1983 (peaked at 37.1%). However, this interest has been on the decline since.
* Women representation in Engineering has been steadily increasing since 1970 (up from less than 1%!). However, this momentum has slowed down since around 1998.
* We see that female representation has increased by around 40% in Agriculture and Business since 1970
* Architecture, Psycology and Biology see a 30% increase in female representation. 
* Foreign Languages, while still predominantly female, sees a 4.4% drop (73.8% to 69.5%). 
* Computer Science and Math rank as the slowest growing STEM degrees amonng women. (Around 4%)
* Most degrees with female representation growing at a faster rate than the overall average are STEM degrees.
* Computer Science is growing at only a quarter of the average pace, Engineering seems to be growing at the average pace.

### Dataset
The dataset used was compiled from Chapter 3 (tables 325.10 through 325.95) of the 2013 version of the Digest of Education Statistics. The Digest includes a selection of data from many sources, both government and private, and draws especially on the results of surveys and activities carried out by the National Center for Education Statistics (NCES). This dataset contains statistics of the percentage of Bachelor’s degrees awarded to women from 1970 to 2011. The data is recorded every year for 17 different fields.
More data can be found on the [National Center for Education Statistics](https://nces.ed.gov/programs/digest/2013menu_tables.asp) website.

```python
# Import packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read in dataset
df = pd.read_csv('degrees.txt')
df.head()
```

```python
# Seperate degrees based on whether or not they are considered STEM
stem = ['Year', 'Agriculture', 'Architecture', 'Biology',
        'Computer Science', 'Engineering',
        'Health Professions', 'Math and Statistics', 
        'Physical Sciences', 'Psychology']
non_stem = ['Year', 'Art and Performance', 'Business', 
            'Communications and Journalism','Education',
            'English', 'Foreign Languages', 'Public Administration',
            'Social Sciences and History']
df_stem = df[stem]
df_non_stem = df[non_stem]
```

```python
# Create new DataFrame to hold stats across degrees
all_avg = df.drop('Year', axis=1).mean(axis=1)
stem_avg = df_stem.drop('Year', axis=1).mean(axis=1)
non_stem_avg = df_non_stem.drop('Year', axis=1).mean(axis=1)

compare = pd.DataFrame({'Year': list(df.Year), 
                        'All Average': all_avg,
                        'STEM Average': stem_avg,
                        'Non-STEM Average': non_stem_avg,
                        'Computer Science': df['Computer Science'],
                        'Math and Statistics': df['Math and Statistics'],
                        'Engineering': df['Engineering']
                       })
compare.head()
```

```python
pop_1970 = list(df.drop('Year', axis=1).iloc[0])
pop_1990 = list(df.drop('Year', axis=1).iloc[20])
pop_2011 = list(df.drop('Year', axis=1).iloc[41])

pop_1970 = pd.DataFrame({'Degree': list(df.drop('Year', axis=1).columns), 
                        '% Women': pop_1970}).sort_values(by=['% Women'], ascending=False).set_index(np.arange(17))
pop_1990 = pd.DataFrame({'Degree': list(df.drop('Year', axis=1).columns), 
                        '% Women': pop_1990}).sort_values(by=['% Women'], ascending=False).set_index(np.arange(17))
pop_2011 = pd.DataFrame({'Degree': list(df.drop('Year', axis=1).columns), 
                        '% Women': pop_2011}).sort_values(by=['% Women'], ascending=False).set_index(np.arange(17))
```

```python
pop_1970 = pop_1970.iloc[0:6]
pop_1990 = pop_1990.iloc[0:6]
pop_2011 = pop_2011.iloc[0:6]
```

```python
plt.figure(figsize=(35,15))
plt.subplot(1, 3, 1)
degrees1970 = pop_1970['Degree']
degrees1990 = pop_1990['Degree']
degrees2011 = pop_2011['Degree']
women1970 = pop_1970['% Women']
women1990 = pop_1990['% Women']
women2011 = pop_2011['% Women']
barplot1970 = plt.bar(degrees1970, women1970, align='center', alpha=0.5)
barplot1970[0].set_color('m')
plt.title('Top 5 Degrees Awarded to Women in 1970', fontname="Arial", fontsize=30, fontweight='bold')
plt.subplot(1, 3, 2)
barplot1990 = plt.bar(degrees1990, women1990, align='center', alpha=0.5)
barplot1990[0].set_color('m')
barplot1990[3].set_color('m')
plt.title('Top 5 Degrees Awarded to Women in 1990', fontname="Arial", fontsize=30, fontweight='bold')
plt.subplot(1, 3, 3)
barplot2011 = plt.bar(degrees2011, women2011, align='center', alpha=0.5)
barplot2011[0].set_color('m')
barplot2011[3].set_color('m')
plt.title('Top 5 Degrees Awarded to Women in 2011', fontname="Arial", fontsize=30, fontweight='bold')
plt.tight_layout()
plt.savefig('top_5_degrees.png')
plt.show()
```

## Top 5 most popular degrees among women 
Health professions remains the most popular degree among women all through the years.
We see Psychology entering the top 5 around 1990.

```python
pop_1970_stem = pop_1970_stem.iloc[0:6]
pop_1990_stem = pop_1990_stem.iloc[0:6]
pop_2011_stem = pop_2011_stem.iloc[0:6]
```

```python
plt.figure(figsize=(35,15))
plt.subplot(1, 3, 1)
stem_degrees1970 = pop_1970_stem['STEM Degree']
stem_degrees1990 = pop_1990_stem['STEM Degree']
stem_degrees2011 = pop_2011_stem['STEM Degree']
stem_women1970 = pop_1970_stem['% Women']
stem_women1990 = pop_1990_stem['% Women']
stem_women2011 = pop_2011_stem['% Women']
stem_barplot1970 = plt.bar(stem_degrees1970 , stem_women1970, align='center', alpha=0.5)
stem_barplot1970[5].set_color('#DB7F67')
plt.title('Top 5 STEM Degrees Awarded to Women in 1970', fontname="Arial", fontsize=30, fontweight='bold')
plt.subplot(1, 3, 2)
stem_barplot1990 = plt.bar(stem_degrees1990 , stem_women1990, align='center', alpha=0.5)
stem_barplot1990[3].set_color('#CE84AD')
plt.title('Top 5 STEM Degrees Awarded to Women in 1990', fontname="Arial", fontsize=30, fontweight='bold')
plt.subplot(1, 3, 3)
stem_barplot2011 = plt.bar(stem_degrees2011 , stem_women2011, align='center', alpha=0.5)
stem_barplot2011[4].set_color('#CE84AD')
plt.title('Top 5 STEM Degrees Awarded to Women in 2011', fontname="Arial", fontsize=30, fontweight='bold')
plt.tight_layout()
plt.savefig('top_5_STEM_degrees.png')
plt.show()
```

## Top 5 most popular STEM degrees among women 
Health Professions, Psychology and Biology remain the top 3 STEM degrees for women. While computer science was the 6th most popular degree (13.6%), it is replaced by other STEM degrees (such as Math, Architecture and Agriculture) in subsequenct years. 

```python
# Compare trends in computer science, math and statistics against the average
years = df['Year']
compare = compare.drop('Year', axis=1)
c = compare.columns
plt.figure(figsize=(10,10))
for x in c:
    plt.plot(years, compare[x], label=x)
    
plt.ylabel('% of Degrees Awarded', fontname="Arial", fontsize=20, fontweight='bold')
plt.xlabel('Year', fontname="Arial", fontsize=20, fontweight='bold')
plt.title('% of Degrees Awarded to Women', fontname="Arial", fontsize=30, fontweight='bold')    
plt.legend(loc=4)
plt.savefig('degrees_women.png')
plt.show
```

# Overall Change in % of Degree Awarded to Women per Degree Type
Now let's look at which field have had the largest changes in female representation since 1970!

```python
# Create new DataFrame to hold the difference for each degree between 1970 and 2011
num_of_deg_names = list(range(1, 18))
diff_list = []
for n in num_of_deg_names:
    diff = (df.iloc[-1,n] - df.iloc[1,n])
    diff_list.append(diff)

t = {'Degree': list(df_2.columns), 'Change (%)': diff_list}

test = pd.DataFrame(t)
average_change = test['Change (%)'].mean(axis=0)
stat = pd.DataFrame({'Degree': ['Average'], 'Change (%)': [average_change]})
test = test.append(stat)
test = test.sort_values(by=['Change (%)'], ascending=False).set_index(np.arange(18))
test
```

We see that female representation has increased by around 40% in Agriculture and Business; while Architecture, Psycology and Biology see a 30% increase in female representation. Foreign Languages, while still predominantly female, sees a 4.4% drop (73.8% to 69.5%). Computer Science and Math rank as the slowest growing STEM degrees amonng women. The bar graph below shows the difference in female representation for each degree type.

```python
# Vizualize the difference for each degree between 1970 and 2011
plt.figure(figsize=(35,35))
plt.axes([0.05, 0.05, 0.5, 0.6])
degrees = test['Degree']
change = test['Change (%)']
barplot = plt.bar(degrees, change, align='center', alpha=0.5)
barplot[7].set_color('g')
barplot[0].set_color('m')
barplot[2].set_color('m')
barplot[3].set_color('m')
barplot[4].set_color('m')
barplot[6].set_color('m')
barplot[8].set_color('m')
barplot[11].set_color('m')
barplot[13].set_color('m')
barplot[14].set_color('m')
plt.ylabel('Change in % of Degree Awarded to Women', fontname="Arial", fontsize=40, fontweight='bold')
plt.xlabel('Degrees', fontname="Arial", fontsize=40, fontweight='bold')
plt.tick_params(labelsize=30)
plt.xticks(rotation=90)
plt.title('Change in % of Degree Awarded to Women between 1970 and 2011', fontname="Arial", fontsize=40, fontweight='bold')
plt.savefig('degrees_change.png', bbox_inches='tight')
plt.show()
```

It is delightful to see that most degrees with female representation growing at a faster rate than the overall average are STEM degrees. While Computer Science is growing at only a quarter of the average pace, Engineering seems to be growing at the average pace.
