# Analysis of the classic Iris dataset

![](img/Iris.jpg)


### Objective
Predict flower species based on 4 features: sepal length and width, and petal length and width

### Dataset
* Iris data set

Source: [UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/iris)

```python
# Imports packages
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Set plotting style
plt.style.use('ggplot')
```

```python
# Load dataset
iris = datasets.load_iris()
```

```python
# Checking type and investigate dataset attributes
print('The dataset type is', type(iris), '\n')
print('The dataset keys are:')
for key in iris.keys():
    print(key)
print('\n') 

print('The feature data type and target data type are', type(iris.data), 'and', type(iris.target), '\n')
print('There are', iris.data.shape[0], 'rows and', iris.data.shape[1], 'columns')
```

Bunch is similar to a dictionary in that it contains key value pairs. We have 150 samples and 4 features: petal length and width and sepal length and width. The target variables are encoded as 0 for 'setosa', 1 for 'versicolor' and 2 for 'virginica'.

```python
# Assign feature and target data to X and Y respectively
x = iris.data
y = iris.target
# Build a DataFrame with feature data and also pass column names
df = pd.DataFrame(x, columns=iris.feature_names)
```

```python
df.head()
```

```python
# Use pandas scatter matrix to visualize dataset
# Color data points by species
# Specify maker size and shape
sm = pd.scatter_matrix(df, c = y, figsize = [8, 8], 
                      s=150, marker = 'D')
```

On the diagonal, we have histograms of the features corresponding to the row and column. The off-diagonal figures are scatter plots of the column feature vs ro feature colored by the target variable (species). We see that petal width and length are highly correlated and flowers are clustered according to species.

We want to build a classifier that takes unlabeled data as input and outputs a label. First we need to choose a type of classifier and it needs to learn from the already labeled data (training data).

Let's go with k-nearest Neighbors.
First let's visually inspect the relationship between petal width, petal length, and the specie of the flower.

```python
target = pd.Series(y)
```

```python
colors = target.map({0: '#728EB1', 1: '#A64253', 2: '#4AF2A1'})
ped_lw = df.plot.scatter('petal length (cm)', 'petal width (cm)', c= colors)
```

```python
# Using scikit-learn to fit a classifier
knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(iris['data'], iris['target'])
```

```python
# Let's try predicting the following set of unlabeled data
x_new = np.array([[4, 3, 3, 3],
       [5, 3, 4, 2],
       [5, 4, 1, 0.5]])
x_new.shape
```

```python
prediction = knn.predict(x_new)
print('Prediction {}'.format(prediction))
```

The predicted labels are versicolor, versicolor, setosa respectively.
